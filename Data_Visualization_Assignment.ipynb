```python
#       -:BOKEH ASSIGNMENT:-
#Q1:  Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10
#     and y-values as the sine of x.
#Answer:
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # Sine of x

# Output to notebook (or you can use output_file('sine_wave.html') for an HTML file)
output_notebook()

# Create a Bokeh figure
p = figure(title='Sine Wave', x_axis_label='X', y_axis_label='Sine of X',
           plot_width=800, plot_height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='sin(x)', line_width=2, color='blue')

# Show the plot
show(p)


#Q2:  Create a Bokeh scatter plot using randomly generated x and y values.
#     Use different sizes and colors for the markers based on the 'sizes' and
#    'colors' columns.

#Answer:
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(10, 100, size=n)  # Random sizes for the markers
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)  # Random colors

# Create a DataFrame
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Output to notebook
output_notebook()

# Create a Bokeh figure
p = figure(title='Random Scatter Plot', x_axis_label='X', y_axis_label='Y',
           plot_width=800, plot_height=400)

# Add a scatter renderer with variable sizes and colors
p.scatter('x', 'y', size='sizes', color='colors', source=data, alpha=0.6)

# Show the plot
show(p)



#Q3:  Generate a Bokeh bar chart representing the counts of different fruits
#     using the following dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]


from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Define the dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Prepare the data
data = {'fruits': fruits, 'counts': counts}

# Output to notebook
output_notebook()

# Create a Bokeh figure
p = figure(x_range=data['fruits'], title='Fruit Counts',
           x_axis_label='Fruits', y_axis_label='Counts',
           plot_width=600, plot_height=400)

# Add bars to the figure
p.vbar(x=data['fruits'], top=data['counts'], width=0.5, color='blue')

# Show the plot
show(p)


#Q4:   Create a Bokeh histogram to visualize the distribution of the given data.
data_hist = np.random.randn(1000)

# Create a histogram using Bokeh
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
p = figure(title="Histogram Example", x_axis_label="Value", y_axis_label="Frequency")

# Add the histogram glyph to the figure




#Q5:  Create a Bokeh heatmap using the provided dataset.
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)


import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
from bokeh.models import ColorBar, LinearColorMapper

# Create the heatmap data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)

# Create a meshgrid for the heatmap
xx, yy = np.meshgrid(x, y)

# Prepare the data for Bokeh
heatmap_data = {
    'x': xx.flatten(),
    'y': yy.flatten(),
    'value': data_heatmap.flatten()
}

# Output to notebook
output_notebook()

# Create a color mapper
mapper = LinearColorMapper(palette=Viridis256, low=data_heatmap.min(), high=data_heatmap.max())

# Create a Bokeh figure
p = figure(title='Heatmap', x_axis_label='X', y_axis_label='Y',
           plot_width=600, plot_height=600)

# Add a rectangle glyph to the figure
p.rect(x='x', y='y', width=0.1, height=0.1, source=heatmap_data,
        line_color=None, fill_color=linear_cmap('value', Viridis256, low=data_heatmap.min(), high=data_heatmap.max()))

# Add a color bar
color_bar = ColorBar(color_mapper=mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)


```
