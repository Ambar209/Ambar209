```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

# CSV file ko read karna
df = pd.read_csv("filename.csv")  # change filename
df.head()


import pandas as pd

# CSV file ko read karna
df = pd.read_csv("filename.csv")  # change filename
df.head()


import matplotlib.pyplot as plt

df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title('Sales by Category')
plt.ylabel('Total Sales')
plt.show()


import plotly.express as px

fig = px.pie(df, names='Region', values='Sales', title='Sales by Region')
fig.show()


import plotly.express as px

fig = px.pie(df, names='Region', values='Sales', title='Sales by Region')
fig.show()


import plotly.express as px
fig = px.bar(df, x='Category', y='Sales', color='Sub-Category', barmode='group')
fig.show()


import pandas as pd

# Replace with your actual file name
df = pd.read_csv('web_series.csv')
df.head()

# Group by release year and count entries
series_per_year = df['Release_Year'].value_counts().sort_index()

# Optional: If Release_Year is not integer
series_per_year.index = series_per_year.index.astype(int)

# Convert to DataFrame for plotting
series_per_year_df = series_per_year.reset_index()
series_per_year_df.columns = ['Year', 'Number of Web Series']

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(series_per_year_df['Year'], series_per_year_df['Number of Web Series'], marker='o', linestyle='-')
plt.title('Trend of Web Series Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Web Series')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


import plotly.express as px

fig = px.line(series_per_year_df, x='Year', y='Number of Web Series',
              title='Trend of Web Series Released Per Year',
              markers=True)
fig.show()


# Task 1: Trend of Web Series Released Per Year

# Task 2: Distribution of IMDB Ratings Across Streaming Platforms


# Rename for consistency (if needed)
df.rename(columns={'IMDB_Rating': 'Rating', 'Platform': 'Platform'}, inplace=True)

# Drop rows with missing values in Rating or Platform
df = df.dropna(subset=['Rating', 'Platform'])

# Ensure 'Rating' is numeric
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')


import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Platform', y='Rating')
plt.title('Distribution of IMDB Ratings Across Streaming Platforms')
plt.xlabel('Streaming Platform')
plt.ylabel('IMDB Rating')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()


```
