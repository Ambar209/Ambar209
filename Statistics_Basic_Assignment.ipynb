Q2:  What are the measures of central tendency, and when should you use each?
     Discuss the mean, median,and mode with examples and situations where each is appropriate

Answer: Central tendency refers to a value that represents the center or  
        typical value of a data set. The three main measures are mean, median, and mode

        Mean (Average)
        The mean is the sum of all values divided by the number of values.

        Use the mean when the data is symmetrical (no extreme values) and when you want a value that considers all data points.

        Median (Middle Value)
        The median is the middle value when the data is arranged in order (either ascending or descending).

        Use the median when the data has outliers (extreme values) or is skewed. The median is not affected by very high or low values.

        Mode (Most Frequent Value)
        The mode is the value that appears most frequently in the data set.
        Use the mode for categorical or nominal data where you want to find the most common category or value.

Q3:  Explain the concept of dispersion. How do variance and standard deviation
     measure the spread of data

Answer: Dispersion refers to how spread out or scattered the values in a data
        set are. It helps to understand the variability or consistency of the data. If all values are similar, the dispersion is low. If values are very different, the dispersion is high.

        Variance
        Variance is the average of the squared differences between each data point and the mean. It measures the spread of data points in relation to the mean

        Standard Deviation
        The standard deviation is the square root of the variance. It is the most commonly used measure of dispersion because it is in the same units as the original data, making it easier to interpret.

Q4: What is a box plot, and what can it tell you about the distribution of data

Answer: A box plot is a graphical representation that summarizes the
        distribution of a data set. It shows the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum values, helping to understand the spread and skewness of the data
        Box: Represents the middle 50% of the data, between Q1 (25th percentile) and Q3 (75th percentile).
        Median (Q2): A line inside the box showing the middle value (50th percentile).
        Whiskers: Lines extending from the box to the minimum and maximum values (excluding outliers).

Q8: Discuss the conditions under which the binomial distribution is used.

Answer: The binomial distribution is used to model situations where there are a
        fixed number of trials, and each trial has only two possible outcomes (success or failure). The conditions for using a binomial distribution are:
        Fixed Number of Trials: The number of trials (n) is set in advance and does not change.

        Two Outcomes: Each trial has only two possible outcomes: success (e.g., heads) or failure (e.g., tails).

        Constant Probability: The probability of success (p) remains the same for each trial.

        Independent Trials: The outcome of one trial does not affect the outcome of other trials.

        Symmetry: The normal distribution is perfectly symmetrical around the mean. This means the left side is a mirror image of the right side.

        Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all the same and lie at the center of the distribution.

        Bell-Shaped Curve: The curve is highest at the mean and tapers off symmetrically as you move away from the center.


       Properties of the Normal Distribution

      The normal distribution is a bell-shaped, symmetrical distribution that is commonly used in statistics. Here are its key properties:

      Symmetry: The normal distribution is perfectly symmetrical around the mean. This means the left side is a mirror image of the right side.

      Mean, Median, and Mode: In a normal distribution, the mean, median, and mode are all the same and lie at the center of the distribution.

      Bell-Shaped Curve: The curve is highest at the mean and tapers off symmetrically as you move away from the center.

      Asymptotic: The tails of the normal distribution curve approach, but never actually touch, the horizontal axis.

      Defined by Mean and Standard Deviation: The normal distribution is fully described by two parameters:

      Mean : The center of the distribution.

      Standard Deviation : Determines the spread or width of the curve. A larger standard deviation means a wider curve.
