Q1: What is Simple Linear Regression

Answer:Simple Linear Regression is a way to predict one thing (like sales)  
       using another thing (like advertising budget) by drawing a straight line through the data.

Q2: What are the key assumptions of Simple Linear Regression

Answer: Linearity: The relationship between the two variables is a straight
        line.
        Independence: The data points are independent of each other.

Q3: - What does the coefficient m represent in the equation Y=mX+c

Answer: In the equation Y = mX + c, the coefficient m represents the slope of
        the line.

Q4: What does the intercept c represent in the equation Y=mX+c

Answer: In the equation Y = mX + c, the intercept c represents the point
        where  the line crosses the Y-axis.

Q5: How do we calculate the slope m in Simple Linear Regression

Answer:  you find how much X and Y change together and divide it by how much X
         changes by itself. This gives you the slope

Q6  What is the purpose of the least squares method in Simple Linear Regression

Answer: least squares method is used to find the best-fitting line in
        simple   linear regression

Q7: How is the coefficient of determination (R²) interpreted in Simple Linear
    Regression

Answer:  the coefficient of determination (R²) tells you how much of the
         variation in the dependent variable (Y) can be explained by the independent variable (X) in a simple linear regression mode

Q8:  What is Multiple Linear Regression

Answer: Multiple linear regression is a method used to predict a dependent
        variable (Y) based on two or more independent variables

Q9: What is the main difference between Simple and Multiple Linear Regression

Answer: The main difference is that simple linear regression uses one
        independent variable to predict the dependent variable, while multiple linear regression uses two or more independent variables to make the prediction

Q10: What are the key assumptions of Multiple Linear Regression

Answer: Linearity: The relationship between the independent variables and the
        dependent variable is linear.

        Independence: The observations (data points) are independent of each other

Q11: What is heteroscedasticity, and how does it affect the results of
     a Multiple Linear Regression model

Answer: it means that the spread of the data gets wider or narrower as the
        values of the predictors change.

Q12: How can you improve a Multiple Linear Regression model with high
     multicollinearity

Answer: Remove one of the correlated variables: If two predictors are highly
        related, keep only one.


       Combine correlated variables: Create a new variable that combines them, like taking an average.

Q13: What are some common techniques for transforming categorical variables for
     use in regression models

Answer:One-hot encoding: Converts each category into a new binary (0 or 1)
       column, where 1 represents the presence of the category and 0 its absence.

       Label encoding: Assigns a unique integer to each category (useful for ordinal categories with a natural order).

       Binary encoding: A mix of one-hot and label encoding, converting categories into binary numbers.

Q14: What is the role of interaction terms in Multiple Linear Regression

Answer: Interaction terms in multiple linear regression show how two or more
        independent variables work together to affect the dependent variable

Q15:  How can the interpretation of intercept differ between Simple and  
      Multiple Linear Regression

Answer: simple linear regression, the intercept represents the value of the
        dependent variable when the independent variable is 0

        multiple linear regression, the intercept represents the value of the dependent variable when all the independent variables are 0

Q16: What is the significance of the slope in regression analysis, and how does
     it affect predictions

Answer: The slope in regression analysis shows how much the dependent variable
        (what you're predicting) changes when the independent variable (the factor you're using for prediction) changes

Q17:How does the intercept in a regression model provide context for the
     relationship between variables

Answer: The intercept in a regression model shows the predicted value of the
        dependent variable when the independent variable is zero. It gives a starting point or baseline for the relationship

Q18: What are the limitations of using R² as a sole measure of model performance

Answer: R² shows how well the model explains the data, but it has limitations.
        It doesn't tell you if the model is the best fit, just how much of the variation is explained

Q19:  How would you interpret a large standard error for a regression
      coefficient

Answer:A large standard error for a regression coefficient means that the
       estimate of the coefficient is uncertain or imprecise

Q20: How can heteroscedasticity be identified in residual plots, and why is it
     important to address it

Answer: Heteroscedasticity can be identified in residual plots if the spread of
        the residuals (errors) changes as the value of the independent variable increases.

Q21: What does it mean if a Multiple Linear Regression model has a high R² but
     low adjusted R²

Answer: If a Multiple Linear Regression model has a high R² but low adjusted  
        R², it means the model explains a lot of the data, but it may be overfitting. The high R² might come from adding too many variables, even if they're not helpful

Q22:Why is it important to scale variables in Multiple Linear Regression

Answer: caling variables in Multiple Linear Regression is important because it
        makes sure all variables are on the same scale, which helps the model treat them equally

Q23: What is polynomial regression

Answer: Polynomial regression is a type of regression where the relationship
        between the independent and dependent variables is modeled as a polynomial (like a curved line), instead of a straight line

Q24: How does polynomial regression differ from linear regression

Answer: Polynomial regression differs from linear regression in that it models
        the relationship between variables as a curve (using powers of the independent variable), while linear regression models it as a straight line
